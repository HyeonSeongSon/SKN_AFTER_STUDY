```python
from langchain_openai_retest import LabelRetest
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_excel(r'C:\Users\user\Desktop\SKN_AFTER_STUDY\data\33증강데이터_48개.xlsx')
df.head()
```

```python
df['re_category1']=''
df['re_category2']=''
```

```python
retest = LabelRetest()
for i in range(213,len(df)):
  input_data = df.iloc[i,:]
  response = retest.create_categories(input_data['generator_context'], input_data['category1'], input_data['category2'])
  categories = response.split(',')
  df.loc[i,'re_category1']=categories[0]
  df.loc[i,'re_category2']=categories[1]
  time.sleep(1)
```
